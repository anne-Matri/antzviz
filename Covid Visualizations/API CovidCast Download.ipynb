## pulling data by state for the following signals

safegraph //
    completely_home_prop /
    full_time_work_prop /

```python
move_home = covidcast.signal("safegraph", "completely_home_prop", 
                         date(yr, strmon, strdt), date(yr, mon, enddt),
           `             geo_type="state")
```

```python
move_fullw = covidcast.signal("safegraph", "full_time_work_prop", 
                        date(yr, strmon, strdt), date(yr, mon, enddt),
                        geo_type="state")
```

```python
move_partw = covidcast.signal("safegraph", "part_time_work_prop", 
                        date(yr, strmon, strdt), date(yr, mon, enddt),
                        geo_type="state")
```

```python
doc_visit = covidcast.signal("doctor-visits", "smoothed_adj_cli", 
                        date(yr, strmon, strdt), date(yr, mon, enddt),
                        geo_type="state")
```

```python
survey_c = covidcast.signal("fb-survey", "smoothed_ili", 
                        date(yr, strmon, strdt), date(yr, mon, enddt),
                        geo_type="state")
```

```python
survey_i = covidcast.signal("fb-survey", "smoothed_cli",
                        date(yr, strmon, strdt), date(yr, mon, enddt),
                        geo_type="state")
```

```python
hv = covidcast.signal("hospital-admissions", "smoothed_covid19_from_claims",
                        date(yr, strmon, strdt), date(yr, mon, enddt),
                        geo_type="state")
```

```python
c_tests = covidcast.signal("quidel", "covid_ag_smoothed_pct_positive",
                          date(yr, strmon, strdt), date(yr, mon, enddt),
                        geo_type="state")
```

```python
search = covidcast.signal("ght", "smoothed_search",
                         date(yr, strmon, strdt), date(yr, mon, enddt),
                        geo_type="state")
```

```python
jhu_case = covidcast.signal("jhu-csse", "confirmed_incidence_prop",
                        date(yr, strmon, strdt), date(yr, mon, enddt),
                        geo_type="state")
```

```python
jhu_deaths_100 = covidcast.signal("jhu-csse", "deaths_incidence_prop",
                        date(yr, strmon, strdt), date(yr, mon, enddt),
                        geo_type="state")
jhu_deaths_abs = covidcast.signal("jhu_case", "confirmed_incidence_num",
                        date(yr, strmon, strdt), date(yr, mon, enddt),
                        geo_type='state')
```

```python
move_home.rename(columns={'value': 'move_home'}, inplace=True)
move_fullw.rename(columns={'value': 'move_fullw'}, inplace=True)
move_partw.rename(columns={'value': 'move_partw'}, inplace=True)
search.rename(columns={'value': 'search'}, inplace=True)
doc_visit.rename(columns={'value': 'doc_visit'}, inplace=True)
survey_c.rename(columns={'value': 'survey_c'}, inplace=True)
survey_i.rename(columns={'value': 'survey_i'}, inplace=True)
c_tests.rename(columns={'value': 'c_tests'}, inplace=True)
hv.rename(columns={'value': 'hv'}, inplace=True)
jhu_case.rename(columns={'value': 'jhu_case'}, inplace=True)
jhu_deaths.rename(columns={'value': 'jhu_d'}, inplace=True)
```

```python
sub1 = move_home[["geo_value", "time_value", "move_home"]]
sub2 = move_fullw[["geo_value", "time_value", "move_fullw"]]
```

```python
sub3 = move_partw[["geo_value", "time_value", "move_partw"]]
sub4 = search[["geo_value", "time_value", "search"]]
sub5 = doc_visit[["geo_value", "time_value", "doc_visit"]]
sub6 = survey_c[["geo_value", "time_value", "survey_c"]]
sub7 = survey_i[["geo_value", "time_value", "survey_i"]]
sub8 = c_tests[["geo_value", "time_value", "c_tests"]]
sub9 = hv[["geo_value", "time_value", "hv"]]
sub10 = jhu_case[["geo_value", "time_value", "jhu_case"]]
sub11 = jhu_deaths[["geo_value", "time_value", "jhu_d"]]
```

```python
move_full = sub1.merge(sub2, how='outer', left_on = ["geo_value", "time_value"], right_on = ["geo_value", "time_value"])
```

```python
move_full = move_full.merge(sub3, how = 'outer', left_on = ["geo_value", "time_value"], right_on = ["geo_value", "time_value"])
```

```python
sub45 = sub4.merge(sub5, how = 'outer', left_on = ["geo_value", "time_value"], right_on = ["geo_value", "time_value"])
```

```python
sub67 = sub6.merge(sub7, how = 'outer', left_on = ["geo_value", "time_value"], right_on = ["geo_value", "time_value"])
```

```python
sub89 = sub8.merge(sub9, how = 'outer', left_on = ["geo_value", "time_value"], right_on = ["geo_value", "time_value"])
```

```python
sub1011 = sub10.merge(sub11, how = 'outer', left_on = ["geo_value", "time_value"], right_on = ["geo_value", "time_value"])
```

```python
sub47 = sub45.merge(sub67, how = 'outer', left_on = ["geo_value", "time_value"], right_on = ["geo_value", "time_value"])
```

```python
sub811 = sub89.merge(sub1011, how = 'outer', left_on = ["geo_value", "time_value"], right_on = ["geo_value", "time_value"])
```

```python
sub411 = sub47.merge(sub811, how = 'outer', left_on = ["geo_value", "time_value"], right_on = ["geo_value", "time_value"])
```

```python
full_symp = sub411.merge(move_full, how = 'outer', left_on = ["geo_value", "time_value"], right_on = ["geo_value", "time_value"])
```

```python
full_symp.fillna(0, inplace=True)
```

```python
statell = pd.read_csv('state_ll.csv')
statell = statell.drop(columns = ["state"])
```

```python
full_symp = full_symp.merge(statell, how = 'outer', left_on='geo_value', right_on='state_lower')
```
